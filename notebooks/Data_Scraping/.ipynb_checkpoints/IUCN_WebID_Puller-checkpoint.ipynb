```python
# Dependency
import pandas as pd
```

```python
# Read animal species data
animal_df = pd.read_csv("../../data/animal_list_original.csv", encoding="UTF-8")

# Preview "df"
animal_df.head()
```

```python
# Delete "Other_Name" column from "animal_df"
del animal_df['Other_Name']

# Preview "animal_df"
animal_df.head()
```

```python
# list to store animal web ID
url_ids = []

# Loop through all full length url for animals on www.iucnredlist.org
for i in range(len(animal_df)):
    # Only retrieve animal web ID from full length url and append it to "url_ids"
    url_id = str(animal_df['Full_url'][i].split("/")[4]) + "/" + \
        str(animal_df['Full_url'][i].split("/")[5])
    
    url_ids.append(url_id)
    
# View "url_ids"
url_ids
```

```python
# Add "IUCN_WebID" column in "animal_df" to store url_ids data
animal_df["IUCN_WebID"] = url_ids

# Preview "animal_df"
animal_df.head()
```

```python
# Delete "Full_url" column from "animal_df"
del animal_df['Full_url']

# Preview "animal_df"
animal_df.head()
```

```python
# Save df as csv file
animal_df.to_csv("../../data/animal_iucn_webid.csv", index=False, header=True)
```
